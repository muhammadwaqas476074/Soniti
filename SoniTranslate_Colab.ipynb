```python
# @title Install requirements for SoniTranslate
%cd /content
!git clone https://github.com/r3gm/SoniTranslate.git
%cd SoniTranslate

# Clean conflicting Colab packages
!pip uninstall chex pandas-stubs ibis-framework albumentations albucore jax -y -q

# Install uv package manager
!pip install uv==0.8.13 -q
!uv venv --python 3.10 --clear -q

# Pre-install build tools
!uv pip install --python .venv -q setuptools==70.0.0 wheel

# System dependencies
!sudo apt-get update -qq
!sudo apt-get install -y -qq --no-install-recommends git-lfs ffmpeg libavcodec-extra
!git lfs install --skip-smudge

# Unified install: requirements_custom.txt + overrides.txt
# This handles torch, numpy, faiss, librosa, whisperX, pyannote,
# demucs, TTS, openai, and all sub-dependencies in one pass.
# overrides.txt forces numpy==1.23.5 and librosa==0.9.2 across all transitive deps.
print("📦 Building audio workspace...")
!uv pip install --python .venv --no-build-isolation -q -r requirements_custom.txt -o overrides.txt

# Pre-download Demucs model (Colab network often fails on first run)
print("📦 Pre-downloading Demucs model...")
!mkdir -p /root/.cache/torch/hub/checkpoints
!wget -q -O /root/.cache/torch/hub/checkpoints/f7e0c4bc-ba3fe64a.th \
  https://dl.fbaipublicfiles.com/demucs/hybrid_transformer/f7e0c4bc-ba3fe64a.th || true

# Verification
print("\n✅ Verification:")
!uv run python -c "import whisperx, pyannote.audio, torch, numpy, faiss, librosa, gradio, demucs; print(f'NumPy: {numpy.__version__} | Faiss: {faiss.__version__} | Librosa: {librosa.__version__} | Gradio: {gradio.__version__} | Demucs: {demucs.__version__}'); print('All imports OK!')"

!sudo apt install -y libcudnn8 -q
```
